```python
import numpy as np
import matplotlib.pyplot as plt

# Activation Functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    exps = np.exp(x - np.max(x))  # stability trick
    return exps / np.sum(exps)

# Input range
x = np.linspace(-10, 10, 400)

# Compute outputs
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky = leaky_relu(x)
# For softmax, use a vector (not elementwise)
x_softmax = np.array([1.0, 2.0, 3.0])
y_softmax = softmax(x_softmax)

# Plot
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.plot(x, y_sigmoid, label="Sigmoid", color="blue")
plt.title("Sigmoid")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y_tanh, label="Tanh", color="red")
plt.title("Tanh")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y_relu, label="ReLU", color="green")
plt.plot(x, y_leaky, label="Leaky ReLU", color="orange", linestyle="--")
plt.title("ReLU vs Leaky ReLU")
plt.grid(True)
plt.legend()

plt.subplot(2, 2, 4)
plt.bar(range(len(y_softmax)), y_softmax, color="purple")
plt.title("Softmax (Example Input: [1,2,3])")
plt.xlabel("Class")
plt.ylabel("Probability")

plt.tight_layout()
plt.show()
```
